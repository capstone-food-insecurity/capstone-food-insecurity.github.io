```python
import pytextrank, spacy
import scattertext as st
import pandas as pd
```

```python
dfSample = pd.read_csv('/gdrive/MyDrive/Colab Notebooks/df_data.csv', sep=',')
dfSample.head(10)
```

```python
dfBsamp = dfSample.loc[(dfSample['country'] == 'Burkina Faso')]
dfGsamp = dfSample.loc[(dfSample['country'] == 'Guinea')]
dfMsamp = dfSample.loc[(dfSample['country'] == 'Mali')]
dfNsamp = dfSample.loc[(dfSample['country'] == 'Niger')]
dfSsamp = dfSample.loc[(dfSample['country'] == 'Senegal')]
```

```python
df = dfSample.groupby(['fews_ipc']).sum().transpose()
dfB = dfBsamp.groupby(['fews_ipc']).sum().transpose()
dfG = dfGsamp.groupby(['fews_ipc']).sum().transpose()
dfM = dfMsamp.groupby(['fews_ipc']).sum().transpose()
dfN = dfNsamp.groupby(['fews_ipc']).sum().transpose()
dfS = dfSsamp.groupby(['fews_ipc']).sum().transpose()
df.head(10)
```

```python
df = df.drop(index = ['Unnamed: 0','rain_mean', 'rain_anom', 'ndvi_mean', 'ndvi_anom', 'et_mean', 'et_anom', 'acled_count', 'acled_fatalities', 'fpi','area', 'cropland', 'population', 'ruggedness', 'pasture'])
dfB = dfB.drop(index = ['Unnamed: 0','rain_mean', 'rain_anom', 'ndvi_mean', 'ndvi_anom', 'et_mean', 'et_anom', 'acled_count', 'acled_fatalities', 'fpi','area', 'cropland', 'population', 'ruggedness', 'pasture'])
dfG = dfG.drop(index = ['Unnamed: 0','rain_mean', 'rain_anom', 'ndvi_mean', 'ndvi_anom', 'et_mean', 'et_anom', 'acled_count', 'acled_fatalities', 'fpi','area', 'cropland', 'population', 'ruggedness', 'pasture'])
dfM = dfM.drop(index = ['Unnamed: 0','rain_mean', 'rain_anom', 'ndvi_mean', 'ndvi_anom', 'et_mean', 'et_anom', 'acled_count', 'acled_fatalities', 'fpi','area', 'cropland', 'population', 'ruggedness', 'pasture'])
dfN = dfN.drop(index = ['Unnamed: 0','rain_mean', 'rain_anom', 'ndvi_mean', 'ndvi_anom', 'et_mean', 'et_anom', 'acled_count', 'acled_fatalities', 'fpi','area', 'cropland', 'population', 'ruggedness', 'pasture'])
dfS = dfS.drop(index = ['Unnamed: 0','rain_mean', 'rain_anom', 'ndvi_mean', 'ndvi_anom', 'et_mean', 'et_anom', 'acled_count', 'acled_fatalities', 'fpi','area', 'cropland', 'population', 'ruggedness', 'pasture'])
```

```python
df['High IPC'] = df.iloc[:,[2,3]].sum(axis=1)
df['Low IPC'] = df.iloc[:,[0,1]].sum(axis=1)
dfB['High IPC'] = dfB.iloc[:,[2]].sum(axis=1)
dfB['Low IPC'] = dfB.iloc[:,[0,1]].sum(axis=1)
dfG['High IPC'] = dfG.iloc[:,[1]].sum(axis=1)
dfG['Low IPC'] = dfG.iloc[:,[0]].sum(axis=1)
dfM['High IPC'] = dfM.iloc[:,[2]].sum(axis=1)
dfM['Low IPC'] = dfM.iloc[:,[0,1]].sum(axis=1)
dfN['High IPC'] = dfN.iloc[:,[1,2]].sum(axis=1)
dfN['Low IPC'] = dfN.iloc[:,[0]].sum(axis=1)
dfS['High IPC'] = dfS.iloc[:,[2]].sum(axis=1)
dfS['Low IPC'] = dfS.iloc[:,[0,1]].sum(axis=1)
df.head(10)
```

```python
df = df.drop(df.columns[[0,1,2,3]], axis=1)
dfB = dfB.drop(dfB.columns[[0,1,2]], axis=1)
dfG = dfG.drop(dfG.columns[[0,1]], axis=1)
dfM = dfM.drop(dfM.columns[[0,1,2]], axis=1)
dfN = dfN.drop(dfN.columns[[0,1,2]], axis=1)
dfS = dfS.drop(dfS.columns[[0,1,2]], axis=1)
df.head(10)
```

```python
term_cat_freq = st.TermCategoryFrequencies(df)
term_cat_freqB = st.TermCategoryFrequencies(dfB)
term_cat_freqG = st.TermCategoryFrequencies(dfG)
term_cat_freqM = st.TermCategoryFrequencies(dfM)
term_cat_freqN = st.TermCategoryFrequencies(dfN)
term_cat_freqS = st.TermCategoryFrequencies(dfS)
```

```python
html = st.produce_scattertext_explorer(
	term_cat_freq,
	category='High IPC',
	category_name='High IPC',
	not_category_name='Low IPC',
)
htmlB = st.produce_scattertext_explorer(
	term_cat_freqB,
	category='High IPC',
	category_name='High IPC',
	not_category_name='Low IPC',
)
htmlG = st.produce_scattertext_explorer(
	term_cat_freqG,
	category='High IPC',
	category_name='High IPC',
	not_category_name='Low IPC',
)
htmlM = st.produce_scattertext_explorer(
	term_cat_freqM,
	category='High IPC',
	category_name='High IPC',
	not_category_name='Low IPC',
)
htmlN = st.produce_scattertext_explorer(
	term_cat_freqN,
	category='High IPC',
	category_name='High IPC',
	not_category_name='Low IPC',
)
htmlS = st.produce_scattertext_explorer(
	term_cat_freqS,
	category='High IPC',
	category_name='High IPC',
	not_category_name='Low IPC',
)
```
