```python
import pytextrank, spacy
import scattertext as st
from pprint import pprint
import pandas as pd
```

```python
df = pd.read_csv('/gdrive/MyDrive/Colab Notebooks/food_news_articles/df_data.csv', sep=',')
```

```python
df.head(10)
```

Ensure date is in correct format

```python
df['year_month'] = pd.to_datetime(df['year_month'], format='%Y-%m-%d')
```

```python
df.head(10)
```

Seperate data into two seperate dataframes: before and after covid 19 based on the date of Feb 2020

```python
preCovid_df = df.loc[(df['year_month'] <= '2020-02-01')]
preCovid_df.head(10)
len(preCovid_df)
```

```python
postCovid_df = df.loc[(df['year_month'] > '2020-02-01')]
postCovid_df.head(10)
len(postCovid_df)
```

```python
preCovid_df.insert(1, "Covid", "pre", True)
preCovid_df.head(10)
```

```python
postCovid_df.insert(1, "Covid", "post", True)
postCovid_df.head(10)
```

```python
frames = [preCovid_df, postCovid_df]
  
prepostDF = pd.concat(frames)
len(prepostDF)
```

```python
display(prepostDF)
```

Seperate covid seperated dataframe by each country

```python
BurkinaFasoDF = prepostDF[prepostDF['country'] == 'Burkina Faso']
display(BurkinaFasoDF)
```

```python
MaliDF = prepostDF[prepostDF['country'] == 'Mali']
display(MaliDF)
```

```python
GuineaDF = prepostDF[prepostDF['country'] == 'Guinea']
display(GuineaDF)
```

```python
SenegalDF = prepostDF[prepostDF['country'] == 'Senegal']
display(SenegalDF)
```

```python
NigerDF = prepostDF[prepostDF['country'] == 'Niger']
display(NigerDF)
```

'prepostDF1' serves as the dataframe for all countries terms sorted by term amount

```python
prepostDF1 = prepostDF.groupby(['Covid']).sum().transpose()
display(prepostDF1)
```

```python
prepostDF1 = prepostDF1.drop(index = ['Unnamed: 0','rain_mean', 'rain_anom', 'ndvi_mean', 'ndvi_anom', 'et_mean', 'et_anom', 'acled_count', 'acled_fatalities', 'fpi','area', 'cropland', 'population', 'ruggedness', 'pasture', 'fews_ipc'])
```

```python
display(prepostDF1)
```

Lets look at term frequencies for an individual country... Senegal

```python
SenegalDF1 = SenegalDF.groupby(['Covid']).sum().transpose()
SenegalDF1 = SenegalDF1.drop(index = ['Unnamed: 0','rain_mean', 'rain_anom', 'ndvi_mean', 'ndvi_anom', 'et_mean', 'et_anom', 'acled_count', 'acled_fatalities', 'fpi','area', 'cropland', 'population', 'ruggedness', 'pasture', 'fews_ipc'])
display(SenegalDF1)
```

```python
Senegal_term_cat_freq = st.TermCategoryFrequencies(SenegalDF1)
```

```python
html0 = st.produce_scattertext_explorer(
	Senegal_term_cat_freq,
	category='pre',
	category_name='pre',
	not_category_name='post',
)
```

```python
import IPython
IPython.display.HTML(html0)
```
