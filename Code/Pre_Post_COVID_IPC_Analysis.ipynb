```python
import pytextrank, spacy
import scattertext as st
import pandas as pd
```

```python
df = pd.read_csv('/gdrive/MyDrive/Colab Notebooks/df_data.csv', sep=',')
```

```python
df.head(10)
```

Ensure date is in correct format

```python
df['year_month'] = pd.to_datetime(df['year_month'], format='%Y-%m-%d')
```

```python
df.head(10)
```

Seperate data into two seperate dataframes: before and after covid 19 based on the date of Feb 2020

```python
preCovid_df = df.loc[(df['year_month'] <= '2020-02-01')]
preCovid_df.head(10)
len(preCovid_df)
```

```python
postCovid_df = df.loc[(df['year_month'] > '2020-02-01')]
postCovid_df.head(10)
len(postCovid_df)
```

```python
dfPre = preCovid_df.groupby(['fews_ipc']).sum().transpose()
dfPost = postCovid_df.groupby(['fews_ipc']).sum().transpose()

dfPost.head(10)
```

```python
dfPre = dfPre.drop(index = ['Unnamed: 0','rain_mean', 'rain_anom', 'ndvi_mean', 'ndvi_anom', 'et_mean', 'et_anom', 'acled_count', 'acled_fatalities', 'fpi','area', 'cropland', 'population', 'ruggedness', 'pasture'])
dfPost = dfPost.drop(index = ['Unnamed: 0','rain_mean', 'rain_anom', 'ndvi_mean', 'ndvi_anom', 'et_mean', 'et_anom', 'acled_count', 'acled_fatalities', 'fpi','area', 'cropland', 'population', 'ruggedness', 'pasture'])
```

```python
dfPost.head(10)
```

```python
dfPre['High IPC'] = dfPre.iloc[:,[2]].sum(axis=1)
dfPre['Low IPC'] = dfPre.iloc[:,[0,1]].sum(axis=1)
dfPost['High IPC'] = dfPost.iloc[:,[2,3]].sum(axis=1)
dfPost['Low IPC'] = dfPost.iloc[:,[0,1]].sum(axis=1)
dfPost.head(10)
```

```python
dfPre = dfPre.drop(dfPre.columns[[0,1,2]], axis=1)
dfPost = dfPost.drop(dfPost.columns[[0,1,2,3]], axis=1)
dfPost.head(10)
```

```python
term_cat_freq_pre = st.TermCategoryFrequencies(dfPre)
term_cat_freq_post = st.TermCategoryFrequencies(dfPost)
```

```python
htmlpre = st.produce_scattertext_explorer(
	term_cat_freq_pre,
	category='High IPC',
	category_name='High IPC',
	not_category_name='Low IPC',
)
htmlpost = st.produce_scattertext_explorer(
	term_cat_freq_post,
	category='High IPC',
	category_name='High IPC',
	not_category_name='Low IPC',
)
```
